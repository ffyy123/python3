## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

## 示例：
sum = 0

i = 1

while i <10:

    sum = sum + i
    i = i + 1

## 错误示例：
sum = 0

i = 1

while i <10:

    sum = sum + i

i = i + 1
- 一旦进入死循环可按 Ctrl + c 停止

# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
```

```python
#数字
while 1:
    number=random.randint(1000,9999)
    print('验证码为:',number)
    me = eval(input('<<'))
    if number == me:
        print('验证码输入正确')
        #break
        #print('----')#如果break了，那么之后的都不运行
    else:
        print('验证码输入错误')

```

```python
import random
while 1:
    n1=chr(random.randint(97,122))
    n2=chr(random.randint(97,122))
    n3=chr(random.randint(97,122))
    n4=chr(random.randint(97,122))
    N=n1+n2+n3+n4
    print('验证码是：',N)
    me=input('输入验证码：')
    if me ==N :
        print('验证码正确')
        break
    else:
        print('验证码错误')
```

- 提示用户连续地输入数字直到正确，且提示用户输入的数字是过高还是过低

```python
num1=random.randint(0,10)
while 1:
    num=eval(input('请输入一个数字'))
    if num==num1:
        print('正确')
        break
    elif num<num1:
        print('太小了，请继续猜')
    else:
        print('太大了，请继续猜')
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代
- range(a,b,k)， a，b，k 必须为整数
- a: start
- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

```python
for i in range(10):  #次数
    print(i)
```

```python
for i in range(0,10,2):  #步长为2
    print(i)
```

```python
for i in range(10,0,-1):  #取10到0
    print(i)
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## EP：
- 使用多层循环完成9X9乘法表
- 显示50以内所有的素数

```python
for i in range(1,10):
    for j in range(i,10):
        print("%d*%d=%d"%(j,i,i*j),end='  ')
    print()
```

```python
for i in range(1,10):
    for j in range(1,i+1):
        print("%d*%d=%d"%(j,i,i*j),end='  ')
    print() #换行
```

```python
i=1
while i<=9:
    j=1
    while j<=i:
        a=i*j
        print('%d*%d=%d'%(j,i,a),end=' ')
        j += 1
    print()
    i+=1
```

```python
sum_=0
var=0
list_=[]
list_.extend(eval(input('请输入十个数字：')))       #list.extend向列表中加入多个元素，list.append加入单个元素
for i in list_:
    sum_+=i
mean=sum_/len(list_)
print('平均数：',mean)
for j in list_:
    var+=(j-mean)**2/(len(list_)-1)
deviation=var**(1/2)
print(deviation)
```
